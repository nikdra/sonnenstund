```python
from sonnenstund.models.geosphereaustria.model import StationGeoJSONSerializer
```

```python
import requests
import pandas as pd
from datetime import datetime, date
from shapely.geometry import Point
```

```python
BASE_URL: str = "https://dataset.api.hub.geosphere.at/v1"
```

```python
time_from = datetime.combine(date(2020, 1, 1), datetime.min.time())
time_to = datetime.combine(date(2020, 1, 31), datetime.min.time())
```

```python
station_ids = "105,106,107"
query_url = f"{BASE_URL}/station/historical/klima-v2-1d?station_ids={station_ids}&parameters=so_h,so_h_flag&start={time_from.isoformat()}&end={time_to.isoformat()}"
r = requests.get(query_url)
r.raise_for_status()
response = StationGeoJSONSerializer(**r.json())
```

```python
timestamps = response.timestamps
df = None
for feature in response.features:
    geometry = Point(
        *feature.geometry.coordinates[::-1]
    )  # read coordinates in reverse order
    data = {p.name: p.data for p in feature.properties.parameters.values()}
    d = {"timestamp": timestamps, "geometry": geometry, **data}
    if df is None:
        df = pd.DataFrame(d)
    else:
        df = pd.concat([df, pd.DataFrame(d)], ignore_index=True)
```
